```python
plt.figure(figsize=(8, 5))
sns.boxplot(y='estimated_downloads', data=df)
plt.title('Box Plot of Estimated Downloads')
plt.show()
```

```python
print("Yearly Game Sales Ranked from Highest to Lowest: ")
df["release_date"]=pd.to_datetime(df["release_date"])
df["release_year"]= df["release_date"].dt.year
df.groupby("release_year")["estimated_downloads"].sum().sort_values(ascending=False).head()



```

```python
yearly_downloads = df.groupby("release_year")["estimated_downloads"].sum().reset_index()

plt.figure(figsize=(12,6))
plt.plot(yearly_downloads["release_year"], yearly_downloads["estimated_downloads"], marker='o')

plt.title("Yearly Game Sales")
plt.xlabel("Years")
plt.ylabel("Estimated Download")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
top_games_by_year=df.loc[df.groupby("release_year")["estimated_downloads"].idxmax()][["release_year","game_name","estimated_downloads"]]
print("\nMost Downloaded Game for Each Year:")
top_games_by_year.sort_values("release_year", ascending=False)
```

```python
df["tag_lists"]=df["user_defined_tags"].str.split(", ")
df_exploded=df.explode("tag_lists")
tag_year_grouped=df_exploded.groupby(["release_year", "tag_lists"])["estimated_downloads"].sum().reset_index()
top_tag_per_year=tag_year_grouped.loc[tag_year_grouped.groupby("release_year")["estimated_downloads"].idxmax()]
top_tag_per_year.sort_values("release_year", ascending=False)
```
